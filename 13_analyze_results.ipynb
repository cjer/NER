```python
def get_results_df(crph, corpus, unit, setup):
    dict_res = []
    
    for conf, res, _, _ in crph:
        dr = {}
        for i, r in enumerate(res):
            split_name = chr(ord('a')+i)
            dr['prec_'+split_name], dr['recall_'+split_name], dr['f1_'+split_name] = r
        dr.update(conf)
        dict_res.append(dr)
    rdf = (pd.DataFrame(dict_res)
           .assign(corpus = corpus)
           .assign(setup =  setup)
           .assign(unit = unit)
          )
    pc = [col for col in rdf if col.startswith('prec')]
    rc = [col for col in rdf if col.startswith('recall')]
    fc = [col for col in rdf if col.startswith('f1')]
    rdf = (rdf
           .assign(prec = lambda x: x[pc].sum(axis=1)/len(pc))
           .assign(recall = lambda x: x[rc].sum(axis=1)/len(rc))
           .assign(f1 = lambda x: x[fc].sum(axis=1)/len(fc)))
    return rdf
```

```python
def make_conll_df(path, add_head_stuff=False):
    # CoNLL file is tab delimeted with no quoting
    # quoting=3 is csv.QUOTE_NONE
    df = (pd.read_csv(path, sep='\t', header=None, quoting=3, 
                names = ['ID', 'FORM', 'LEMMA', 'UPOS', 'XPOS', 'FEATS', 'HEAD', 'DEPREL', 'DEPS', 'MISC'])
                # add sentence labels
                .assign(sent = lambda x: (x.ID==1).cumsum())
                # replace bad root dependency tags
                .replace({'DEPREL': {'prd': 'ROOT'}})
               )
    
    if add_head_stuff:
        df = df.merge(df[['ID', 'FORM', 'sent', 'UPOS']].rename(index=str, columns={'FORM': 'head_form', 'UPOS': 'head_upos'}).set_index(['sent', 'ID']),
               left_on=['sent', 'HEAD'], right_index=True, how='left')
    return df
```
